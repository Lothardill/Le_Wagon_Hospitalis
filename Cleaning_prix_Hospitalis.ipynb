```python
import pandas as pd
import numpy as np
```

```python
url = 'C:/Users/lgino/code/Lothardill/Hospitalis/table_clean.csv'
df = pd.read_csv(url)
df.dtypes
```

```python
df['Libellé Article'] = df['Libellé Article'].astype(str)
df["Quantité"] = (df["Quantité"].str.replace(",", ".", regex=False).astype(float))
df['Quantité'] = df['Quantité'].fillna(0)
df["Quantité"] = (df["Quantité"].round(0).astype(int))
df['Dépôt DMI'] = df['Dépôt DMI'].replace("NULL", "0", regex=False)
df['Dépôt DMI'] = df['Dépôt DMI'].fillna(0)
df["Dépôt DMI"] = (df["Dépôt DMI"].astype(int))
```

```python
df.info()
```

```python
# prompt: je veux metre toute mes prix et quantité et mnt de mon df_AA833BB en float

for col in ['Prix HT','Prix TTC', 'Quantité', 'Mnt HT', 'Mnt TTC']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Display the data types to confirm the change
df.dtypes
```

```python
df['prix'] = np.where(
    (df[['Prix HT', 'Prix TTC', 'Mnt HT', 'Mnt TTC']].fillna(0) == 0).all(axis=1),  # Si toutes ces colonnes sont NaN ou 0
    np.NaN,  # On met NaN dans prix
    np.where(pd.notna(df['Prix HT']), df['Prix HT'],
    np.where(pd.notna(df['Prix TTC']), df['Prix TTC'] * 0.8,
    np.where(pd.notna(df['Mnt HT']) & (df['Quantité'] > 0), df['Mnt HT'] / df['Quantité'],
    np.where(pd.notna(df['Mnt TTC']) & (df['Quantité'] > 0), (df['Mnt TTC'] * 0.8) / df['Quantité'],
    np.NaN))  # Sinon, NaN par défaut
    )))
df
```

```python
df.to_csv('table_clean_with_price.csv')
```
