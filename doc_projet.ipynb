```python
import pandas as pd
import glob

# Define your path to the folder with Excel files
folder_path = "/content/drive/MyDrive/Untitled Folder"  # Replace this!

# List all Excel files
import glob
excel_files = glob.glob(f"{folder_path}/*.xlsx")

# Initialize an empty DataFrame list
df_list = []

# Loop over each Excel file and append its contents
for file in excel_files:
    print(f"Reading: {file}")
    temp_df = pd.read_excel(file, engine='openpyxl')
    df_list.append(temp_df)

# Concatenate all DataFrames into one
combined_df = pd.concat(df_list, ignore_index=True)

# Save combined DataFrame as CSV
output_path = "/content/drive/MyDrive/combined_excel_files.csv"
combined_df.to_csv(output_csv, index=False, encoding='utf-8-sig')

print("Combined CSV saved successfully!")
```
