```python
import pandas as pd

# Load Excel file
df = pd.read_excel("C:/Users/lgino/code/Lothardill/Hospitalis/Open LLP 2023.xlsx", sheet_name="Feuil1")

# Define keywords to filter (these can appear anywhere in the string)
keywords = ["AA0B7BB", "AA12WBB ","AA57LBB","AA68BB","AA264BB","AA496","859BB","AAAPPBB","AATERBB","AA13ZBB", "AA35ZBB", "AA57YBB","AA85BB","AA291","AA501BB","AA544BB","A501BB","AAZIMBB","AA832BB","AA01UBB","AA18GBB","AA42HBB","AA44CBB","AA49XBB","AA499BB","AA502BB","AA833BB","AA01WBB","AA16ABB","AA52MBB","AA53GBB","AA62MBB","AA171BB","AA503BB","AA545BB","AA806BB"]  # Add more if needed

# Create a regex pattern from the keywords (join with `|` for "OR" search)
pattern = "|".join(keywords)

# Filter rows where "Étiquettes de lignes" contains any of the keywords
df_filtered = df[df["Étiquettes de lignes"].str.contains(pattern, na=False, case=False)]

# Display filtered rows
print(df_filtered)

```
