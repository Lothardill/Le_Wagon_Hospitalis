```python
def categorize_products_by_supplier(df):
    """
    Catégorise les produits par fournisseur selon :
    - Très intéressant (DMI >= 75% et Prix >= 50)
    - Intéressant (DMI >= 75% OU Prix >= 50)
    - Peu intéressant (le reste)

    Args:
        df: DataFrame contenant les données.

    Returns:
        Un DataFrame avec le fournisseur en index et le nombre de produits par catégorie.
    """
    # Étape 1 : Calcul du ratio DMI par Fournisseur et Réf. Article
    dmi_ratio = df.groupby(['Fournisseur', 'Réf. Article']).agg(
        somme_dmi=('Dépôt DMI', 'sum'),
        count_product=('Dépôt DMI', 'count'),
        prix=('prix', 'mean')  # On prend la moyenne du prix HT par précaution
    )

    dmi_ratio['dmi_ratio'] = dmi_ratio['somme_dmi'] / dmi_ratio['count_product']

    # Étape 2 : Classification des produits
    def classify_product(row):
        if row['dmi_ratio'] >= 0.75 and row['prix'] >= 50:
            return "Très intéressant"
        elif row['dmi_ratio'] >= 0.75 or row['prix'] >= 50:
            return "Intéressant"
        else:
            return "Peu intéressant"

    dmi_ratio['Categorie'] = dmi_ratio.apply(classify_product, axis=1)

    # Étape 3 : Compter les produits par fournisseur et catégorie
    result = dmi_ratio.reset_index().groupby(['Fournisseur', 'Categorie'])['Réf. Article'].nunique().unstack(fill_value=0)

    return result

# Appel de la fonction
categorized_df = categorize_products_by_supplier(df_hospitalis)
categorized_df
```

```python
def categorize_products_by_supplier(df):
    """
    Catégorise les produits par fournisseur selon :
    - Très intéressant (DMI >= 75% et Prix HT >= 50)
    - Intéressant (DMI >= 75% OU Prix HT >= 50)
    - Peu intéressant (le reste)

    Args:
        df: DataFrame contenant les données.

    Returns:
        Le DataFrame avec une nouvelle colonne 'rfid' pour la catégorisation.
    """
    # Calcul du ratio DMI par Fournisseur et Réf. Article
    df['dmi_ratio'] = df.groupby(['Fournisseur', 'Réf. Article'])['Dépôt DMI'].transform('sum') / df.groupby(['Fournisseur', 'Réf. Article'])['Dépôt DMI'].transform('count')

    # Application de la logique pour la classification
    def classify_product(row):
        if row['dmi_ratio'] >= 0.75 and row['prix'] >= 50:
            return "Très intéressant"
        elif row['dmi_ratio'] >= 0.75 or row['prix'] >= 50:
            return "Intéressant"
        else:
            return "Peu intéressant"

    # Créer la colonne 'rfid' avec les catégories
    df['rfid'] = df.apply(classify_product, axis=1)

    # Regrouper par 'Fournisseur' et 'Réf. Article' et appliquer les agrégations
    df_grouped = df.groupby(['Fournisseur', 'Réf. Article']).agg(
            libelle=('Libellé Article', lambda x: x.mode()[0] if not x.mode().empty else None),
            prix=('prix', 'mean'),
            robot_ets=('Robot ETS', 'nunique'),
            date_insertion=('Date Insertion', 'first'),
            quantite=('Quantité', 'sum'),
            dmi_ratio=('dmi_ratio', 'mean'),
            rfid=('rfid', 'first')
            ).reset_index()

    return df_grouped

# Appel de la fonction
categorized_df = categorize_products_by_supplier(df_hospitalis)

# Afficher le DataFrame groupé par 'Fournisseur' et 'Réf. Article' avec les informations souhaitées
categorized_df
```

```python
import pandas as pd

categorized_df = categorized_df[categorized_df["quantite"] != -1]

def normalize_10(series):
    return 10 * (series - series.min()) / (series.max() - series.min())

# Step 1: Normalize the required columns
prix_norm = pd.qcut(categorized_df["prix"], q=10, labels=range(1, 11))
dmi_ratio_norm = normalize_10(categorized_df["dmi_ratio"])
quantite_norm = pd.cut(categorized_df["quantite"], bins=10, labels=range(1, 11))
robot_ets_norm = pd.cut(categorized_df["robot_ets"], bins=10, labels=range(1, 11))

# Step 2: Fill NaN **only** for selected columns

# Step 3: Apply the weighted scoring formula
categorized_df["Scoring_produit"] = (
    (prix_norm.astype(float) * 0.4) +
    (dmi_ratio_norm.astype(float) * 0.4) +
    (robot_ets_norm.astype(float) * 0.1) +
    (quantite_norm.astype(float) * 0.1)
)

# Step 4: Scale "Scoring_produit" between 0 and 10
categorized_df["Scoring_produit"] = normalize(categorized_df["Scoring_produit"]) * 10

categorized_df.sort_values(by=['Scoring_produit'], ascending=False)
```

```python
#1. aggregation fournisseur => count RF => 80 % moyenne scoring produit,  20% nb produit total
#categorized_df["Scoring_fournisseur"] = 
#categorized_df
# Step 1: Compute Aggregations
fournisseur_agg = categorized_df.groupby("Fournisseur").agg(
    avg_scoring_produit=("Scoring_produit", "mean"),  # Average scoring produit per fournisseur
    total_products=("Réf. Article", "nunique")       # Count unique products per fournisseur
).reset_index()

def normalize_10(series):
    return 10 * (series - series.min()) / (series.max() - series.min())


# Step 2: Normalize "total_products" for fair weighting
avg_scoring_produit = normalize_10(fournisseur_agg["avg_scoring_produit"])
total_products_norm = pd.cut(fournisseur_agg["total_products"], bins=10, labels=range(1, 11))

# Step 3: Compute Final "Scoring_fournisseur" (80% avg Scoring_produit, 20% total products)
fournisseur_agg["Scoring_fournisseur"] = (
    (avg_scoring_produit.astype(float) * 0.7) +
    (total_products_norm.astype(float) * 0.3)
)

fournisseur_agg.sort_values(by=['Scoring_fournisseur'], ascending=False)
```
